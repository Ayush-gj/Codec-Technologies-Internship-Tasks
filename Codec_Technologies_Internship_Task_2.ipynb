# **Project 2: Movie Recommendation System**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Generate an inline mock dataset so you don't need to upload any files
movie_data = {
    'movie_id': [0, 1, 2, 3, 4, 5],
    'title': ['The Dark Knight', 'Inception', 'Toy Story', 'Finding Nemo', 'The Conjuring', 'Insidious'],
    'genres': ['Action Crime Drama Thriller', 'Action Sci-Fi Thriller', 'Animation Adventure Comedy', 'Animation Adventure Family', 'Horror Mystery Thriller', 'Horror Mystery']
}

df = pd.DataFrame(movie_data)
print("=== Mini Movie Database ===")
print(df[['title', 'genres']])
print("=" * 27 + "\n")

# 2. Convert genre strings into a token count matrix
vectorizer = CountVectorizer()
count_matrix = vectorizer.fit_transform(df['genres'])

# 3. Calculate the Cosine Similarity scores between all vectors
similarity_scores = cosine_similarity(count_matrix)

# 4. Recommendation Engine Function
def recommend_movies(target_movie_title, standard_recommendation_count=2):
    try:
        # Find the row index belonging to our target movie
        movie_index = df[df['title'].str.lower() == target_movie_title.lower()].index[0]
    except IndexError:
        return f"Error: '{target_movie_title}' is not present in our database."

    # Grab all similarity pairings for this specific movie and sort them descending
    raw_scores = list(enumerate(similarity_scores[movie_index]))
    sorted_scores = sorted(raw_scores, key=lambda x: x[1], reverse=True)

    # Extract index locations of top recommendations (skipping index 0 since that is the movie itself)
    recommended_indices = [item[0] for item in sorted_scores if item[0] != movie_index][:standard_recommendation_count]

    return df['title'].iloc[recommended_indices].tolist()

# Test the system out with an input example
user_choice = "Inception"
results = recommend_movies(user_choice)

print(f"🎬 Because you watched '{user_choice}', you might also like:")
for movie in results:
    print(f"👉 {movie}")
```
