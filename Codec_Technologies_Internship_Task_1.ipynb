# **Project 1: Chatbot for Customer Service**

```python
def customer_chatbot():
    # A dictionary mapping keywords to customer service responses
    responses = {
        "hello": "Hello! Welcome to Codec Support. How can I help you today?",
        "hi": "Hi there! How can I help you with your order or services?",
        "hours": "Our business hours are Monday to Friday, 9:00 AM to 6:00 PM IST.",
        "status": "To check your order or service status, please paste your 8-digit tracking ID.",
        "refund": "We offer a full refund within 14 days of purchase. Would you like me to raise a ticket?",
        "contact": "You can reach our human team at support@codectechnologies.in anytime.",
        "bye": "Thank you for reaching out to us. Have a fantastic day!"
    }

    print("🤖 Chatbot: System initialized. (Type 'bye' to exit the chat)\n")

    while True:
        # Accept live interactive user input inside Google Colab
        user_input = input("You: ").lower().strip()

        if user_input == 'bye' or user_input == 'exit':
            print(f"🤖 Chatbot: {responses['bye']}")
            break

        # Scan user input for keywords
        matched = False
        for keyword, response in responses.items():
            if keyword in user_input:
                print(f"🤖 Chatbot: {response}\n")
                matched = True
                break # Stop matching after finding the primary keyword

        if not matched:
            print("🤖 Chatbot: I didn't quite catch that. Try asking about our 'hours', 'refund' policy, or how to 'contact' support.\n")

# Execute the chatbot
customer_chatbot()
```
